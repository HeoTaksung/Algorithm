### Binary search tree animation
- Make the animation of the creation process of binary search tree using pydot and IPython display
- Before inserting the next node, highlight the nodes you are comparing every second to find the position.
- To highlight the node, change the fontcolor of the node.
- Example node list: [14,57,97,95,74,7,10]

```python
class Node(object):
    def __init__(self, data):
        self.data = data
        self.left = self.right = None
```

```python
class BinarySearchTree(object):
    def __init__(self):
        self.root = None
        
    def insert(self, data):
        self.root = self._insert_value(self.root, data)
        return self.root is not None

    def _insert_value(self, node, data):
        if node is None:
            node = Node(data)
        else:
            if data <= node.data:
                node.left = self._insert_value(node.left, data)
            else:
                node.right = self._insert_value(node.right, data)
        return node

    def mid_order_traversal(self):
        def _mid_order_traversal(root):
            if root is None:
                pass
            else:
                print(root.data)
                _mid_order_traversal(root.left)
                _mid_order_traversal(root.right)
        _mid_order_traversal(self.root)
```

```python
bt=BinarySearchTree()
root=None


li = [14,57,97,95,74,7,10]
bst = BinarySearchTree()
for x in li:
    bst.insert(x)

t = bst.mid_order_traversal()
```

```python
import pydot
from IPython.display import Image, display, update_display
import time
#li = [14,57,97,95,74,7,10]
G=pydot.Dot(graph_type='digraph')

nodea = pydot.Node(bst.root.data, fontcolor='red')
nodeb = pydot.Node(bst.root.left.data, fontcolor='blue')
nodec = pydot.Node(bst.root.right.data, fontcolor='blue')
noded = pydot.Node("None1", style='invis')
nodee = pydot.Node(bst.root.left.right.data, fontcolor='blue')
nodef = pydot.Node("None2", style='invis')
nodeg = pydot.Node(bst.root.right.right.data, fontcolor='blue')
nodeh = pydot.Node(bst.root.right.right.left.data, fontcolor='blue')
nodei = pydot.Node("None3", style='invis')
nodej = pydot.Node(bst.root.right.right.left.left.data, fontcolor='blue')
nodek = pydot.Node("None4", style='invis')



G.add_node(nodea)
G.add_node(nodeb)
G.add_node(nodec)
G.add_node(noded)
G.add_node(nodee)
G.add_node(nodef)
G.add_node(nodeg)
G.add_node(nodeh)
G.add_node(nodei)
G.add_node(nodej)
G.add_node(nodek)


G.add_edge(pydot.Edge(nodea, nodeb))
G.add_edge(pydot.Edge(nodea, nodec))
G.add_edge(pydot.Edge(nodeb, noded))
G.add_edge(pydot.Edge(nodeb, nodee))
G.add_edge(pydot.Edge(nodec, nodef))
G.add_edge(pydot.Edge(nodec, nodeg))
G.add_edge(pydot.Edge(nodeg, nodeh))
G.add_edge(pydot.Edge(nodeg, nodei))
G.add_edge(pydot.Edge(nodeh, nodej))
G.add_edge(pydot.Edge(nodeh, nodek))


plt=Image(G.create_png())
display(plt,display_id=('view'))
for i in range(5):
    if i%2 == 0:
        col = 'red'
    else:
        col = 'green'
    nodea.set_fontcolor(col)
    plt=Image(G.create_png())
    update_display(plt,display_id='view')
    time.sleep(1)
```
