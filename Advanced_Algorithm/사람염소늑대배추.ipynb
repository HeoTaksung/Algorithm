```python
import random                           # 임의의 값을 사용하여 강건너기 알고리즘을 사용

start = ["염소", "늑대", "배추"]         # 강을 건너려면 사람은 무조건 건너야하므로 염소, 늑대, 배추만 사용, 출발지점
thing1 = ["늑대", "염소"]                # 제약조건, sort된 리스트
thing2 = ["배추", "염소"]                # 제약조건, sort된 리스트
end = []                                 # 도착지점

while True:
    if len(start) == 0:                    #무한반복을 종료 (출발지점에 아무것도 없음)
        break
        
    go = random.sample(start,1)            #start에 있는 데이터들 중 하나를 랜덤하게 가져옴
    
    start.remove(go[0])                    #강을 건넜으므로 start에 있는 데이터 제거, end에 추가
    end.append(go[0])
    start.sort()                           # sort를 사용해야 thing1과 thing2를 사용할 수 있음    
    if thing1 == start or thing2 == start:   # 만약 start지점에 제약조건을 가진 데이터들이 있다면 원상복귀
        start.append(go[0])
        end.remove(go[0])
        continue
        
    end.sort()                            #sort를 사용해야 thing1과 thing2를 사용할 수 있음
    print("강 건너기 = ", go[0])
    if thing1 == end or thing2 == end:    # 만약 end지점에 제약조건을 가진 데이터들이 있다면
        print("도착지점 = ", end)
        t = random.sample(end,1)          # end에 있는 데이터들 중 하나를 랜덤하게 가져옴
        end.remove(t[0])                  # 랜덤하게 가져온 데이터 하나를 start지점에 다시 가져다놓음
        start.append(t[0])
        
        print("빼오는 것 = ", t[0])
        print("시작지점 = ", start)
        print("도착지점 = ", end)
        continue
        
    print("시작지점 = ", start)
    print("도착지점 = ", end)

```
